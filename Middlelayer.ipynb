```python
import csv
import json
import os
from datetime import datetime
from IPython.display import display, HTML

# 1. CREATE THE DATA FILE (Safely via Python)
csv_content = """trade_id,timestamp,instrument,trade_type,currency,quantity,entry_price,market_price,status,desk,counterparty
TRD-001,2024-03-14T08:14:22,AAPL,BUY,USD,150,171.50,174.20,SETTLED,Equities,Morgan Stanley
TRD-002,2024-03-14T08:31:05,MSFT,BUY,USD,80,415.00,418.50,SETTLED,Equities,Goldman Sachs
TRD-003,2024-03-14T08:45:17,EUR/SEK,BUY,SEK,500000,11.42,11.38,SETTLED,FX,SEB
TRD-004,2024-03-14T09:02:44,NOVO B,SELL,DKK,200,892.00,901.50,SETTLED,Equities,Nordea
TRD-005,2024-03-14T09:18:33,BTC/EUR,BUY,EUR,2,61200.00,63450.00,SETTLED,Crypto,Coinbase
TRD-006,2024-03-14T09:35:11,USD/EUR,SELL,USD,250000,0.921,0.918,SETTLED,FX,Deutsche Bank
TRD-007,2024-03-14T09:51:28,TSLA,BUY,USD,45,178.30,172.10,SETTLED,Equities,JP Morgan
TRD-008,2024-03-14T10:08:54,GBP/EUR,BUY,GBP,100000,1.172,1.168,SETTLED,FX,Barclays
TRD-009,2024-03-14T10:24:17,SPOT GOLD,BUY,USD,50,2318.00,2341.50,SETTLED,Commodities,HSBC
TRD-010,2024-03-14T10:41:03,VOLV B,BUY,SEK,300,271.80,275.40,SETTLED,Equities,Swedbank
TRD-011,2024-03-14T11:02:39,ETH/EUR,SELL,EUR,15,3210.00,3180.00,SETTLED,Crypto,Kraken
TRD-012,2024-03-14T11:19:22,NOK/EUR,BUY,NOK,750000,0.0862,0.0871,SETTLED,FX,DNB
TRD-013,2024-03-14T11:38:47,NVIDIA,BUY,USD,25,875.00,912.40,SETTLED,Equities,Goldman Sachs
TRD-014,2024-03-14T11:55:14,EUR/PLN,SELL,PLN,400000,4.312,4.298,SETTLED,FX,PKO Bank
TRD-015,2024-03-14T12:14:33,SAMPO,BUY,EUR,180,46.82,47.15,SETTLED,Equities,OP Pohjola
TRD-016,2024-03-14T12:31:08,CHF/EUR,BUY,CHF,80000,1.048,1.051,SETTLED,FX,UBS
TRD-017,2024-03-14T12:48:55,WTI CRUDE,SELL,USD,100,81.40,79.20,SETTLED,Commodities,Vitol
TRD-018,2024-03-14T13:05:22,ERICSSON B,BUY,SEK,500,74.20,73.10,SETTLED,Equities,Handelsbanken
TRD-019,2024-03-14T13:22:41,SOL/EUR,BUY,EUR,80,142.50,138.20,SETTLED,Crypto,Binance
TRD-020,2024-03-14T13:41:18,GBP/USD,SELL,GBP,120000,1.274,1.269,SETTLED,FX,HSBC
TRD-021,2024-03-14T14:02:33,AMAZON,BUY,USD,30,185.40,188.90,SETTLED,Equities,Morgan Stanley
TRD-022,2024-03-14T14:19:07,EUR/NOK,BUY,NOK,600000,11.621,11.598,SETTLED,FX,Nordea
TRD-023,2024-03-14T14:38:44,FORTUM,SELL,EUR,250,14.82,14.55,SETTLED,Equities,OP Pohjola
TRD-024,2024-03-14T14:55:21,XRP/EUR,BUY,EUR,5000,0.582,0.571,SETTLED,Crypto,Coinbase
TRD-025,2024-03-14T15:14:38,SILVER SPOT,BUY,USD,200,27.40,27.85,SETTLED,Commodities,HSBC
TRD-026,2024-03-14T15:31:55,NOKIA,BUY,EUR,800,3.821,3.790,SETTLED,Equities,Nordea
TRD-027,2024-03-14T15:49:12,USD/CHF,SELL,USD,90000,0.895,0.891,SETTLED,FX,Credit Suisse
TRD-028,2024-03-14T16:08:29,VESTAS,BUY,DKK,100,142.50,148.20,SETTLED,Equities,Danske Bank
TRD-029,2024-03-14T16:25:44,BNB/EUR,SELL,EUR,30,392.00,381.50,SETTLED,Crypto,Binance
TRD-030,2024-03-14T16:42:18,NGAS SPOT,BUY,USD,500,2.142,2.198,SETTLED,Commodities,Vitol
TRD-031,2024-03-14T17:01:33,ALPHABET,BUY,USD,20,172.80,175.40,SETTLED,Equities,Goldman Sachs
TRD-032,2024-03-14T17:18:52,SEK/EUR,SELL,SEK,450000,0.0881,0.0878,SETTLED,FX,SEB
TRD-033,2024-03-14T17:38:07,NESTE,BUY,EUR,120,18.42,17.85,SETTLED,Equities,OP Pohjola
TRD-034,2024-03-14T17:55:00,MATIC/EUR,BUY,EUR,2000,0.891,0.842,SETTLED,Crypto,Kraken
TRD-035,2024-03-14T18:14:22,BRENT CRUDE,BUY,USD,80,85.20,86.80,SETTLED,Commodities,Vitol
TRD-036,2024-03-14T22:31:18,TSLA,SELL,USD,60,172.10,164.30,SETTLED,Equities,JP Morgan
TRD-037,2024-03-14T23:08:44,BTC/EUR,SELL,EUR,3,63450.00,68200.00,SETTLED,Crypto,Coinbase
TRD-038,2024-03-14T23:42:11,EUR/USD,BUY,USD,300000,1.086,1.079,SETTLED,FX,Deutsche Bank
TRD-039,2024-03-15T01:15:33,NVIDIA,SELL,USD,30,912.40,891.20,SETTLED,Equities,Goldman Sachs
TRD-040,2024-03-15T02:48:22,ETH/EUR,BUY,EUR,20,3180.00,3420.00,SETTLED,Crypto,Binance
TRD-041,2024-03-14T09:14:55,SPOT GOLD,BUY,USD,100,2318.00,2318.00,FAILED,Commodities,HSBC
TRD-042,2024-03-14T09:52:18,SPOT GOLD,BUY,USD,75,2318.00,2318.00,FAILED,Commodities,HSBC
TRD-043,2024-03-14T10:31:44,SPOT GOLD,BUY,USD,50,2318.00,2318.00,FAILED,Commodities,HSBC
TRD-044,2024-03-14T11:18:22,AAPL,SELL,USD,200,174.20,174.20,FAILED,Equities,Morgan Stanley
TRD-045,2024-03-14T12:44:38,GBP/EUR,SELL,GBP,50000,1.168,1.168,FAILED,FX,Barclays
TRD-046,2024-03-14T08:22:15,PALLADIUM,BUY,USD,30,1042.00,1142.00,SETTLED,Commodities,HSBC
TRD-047,2024-03-14T10:55:33,UNISWAP,BUY,EUR,500,9.82,10.94,SETTLED,Crypto,Kraken
TRD-048,2024-03-14T13:18:44,META,BUY,USD,40,512.00,561.40,SETTLED,Equities,Goldman Sachs
TRD-049,2024-03-14T15:42:11,PLN/EUR,BUY,PLN,800000,0.231,0.248,SETTLED,FX,PKO Bank
TRD-050,2024-03-14T14:08:55,COPPER,SELL,USD,300,4.21,3.89,SETTLED,Commodities,Vitol"""

with open("trades.csv", "w") as f:
    f.write(csv_content)

# 2. MIDDLELAYER CONFIG & LOGIC
FX_RATES_TO_EUR = {
    "EUR": 1.0, "USD": 0.92, "GBP": 1.17, "SEK": 0.088,
    "NOK": 0.086, "DKK": 0.134, "CHF": 1.05, "PLN": 0.23,
}

ANOMALY_RULES = {
    "large_loss_eur": -5000, "large_gain_eur": 8000, "price_deviation_pct": 5.0,
    "late_hour_threshold": 22, "early_hour_threshold": 6, "consecutive_failures": 2,
}

def load_trades(filepath):
    trades = []
    with open(filepath, newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            trades.append({
                "id": row["trade_id"], "timestamp": row["timestamp"], "instrument": row["instrument"],
                "type": row["trade_type"], "currency": row["currency"], "quantity": float(row["quantity"]),
                "entry_price": float(row["entry_price"]), "market_price": float(row["market_price"]),
                "status": row["status"], "desk": row["desk"], "counterparty": row["counterparty"],
            })
    return trades

def calculate_pnl(trade):
    qty, entry, mkt = trade["quantity"], trade["entry_price"], trade["market_price"]
    fx = FX_RATES_TO_EUR.get(trade["currency"], 1.0)
    raw_pnl = ((mkt - entry) * qty) if trade["type"] == "BUY" else ((entry - mkt) * qty)
    pnl_eur = raw_pnl * fx
    price_deviation_pct = abs((mkt - entry) / entry * 100) if entry != 0 else 0
    return {**trade, "pnl_local": round(raw_pnl, 2), "pnl_eur": round(pnl_eur, 2),
            "price_deviation_pct": round(price_deviation_pct, 2), "fx_rate": fx}

def run_pnl_engine(trades):
    return [calculate_pnl(t) if t["status"] == "SETTLED" else {**t, "pnl_local": 0, "pnl_eur": 0, "price_deviation_pct": 0, "fx_rate": 0} for t in trades]

def detect_anomalies(trades):
    flagged = []
    failure_counts = {}
    for t in trades:
        flags = []
        if t["pnl_eur"] < ANOMALY_RULES["large_loss_eur"]: flags.append(f"LARGE_LOSS: €{t['pnl_eur']:,.0f}")
        if t["pnl_eur"] > ANOMALY_RULES["large_gain_eur"]: flags.append(f"LARGE_GAIN: €{t['pnl_eur']:,.0f}")
        if t["price_deviation_pct"] > ANOMALY_RULES["price_deviation_pct"]: flags.append(f"PRICE_DEVIATION: {t['price_deviation_pct']:.1f}%")
        try:
            dt = datetime.fromisoformat(t["timestamp"])
            if dt.hour >= ANOMALY_RULES["late_hour_threshold"] or dt.hour < ANOMALY_RULES["early_hour_threshold"]:
                flags.append(f"OFF_HOURS: {dt.hour:02d}:00")
        except: pass
        if t["status"] == "FAILED":
            key = t["instrument"]
            failure_counts[key] = failure_counts.get(key, 0) + 1
            if failure_counts[key] >= ANOMALY_RULES["consecutive_failures"]: flags.append(f"CONSECUTIVE_FAILURES: {failure_counts[key]}x on {key}")
        if flags: flagged.append({**t, "flags": flags, "flag_count": len(flags)})
    return sorted(flagged, key=lambda x: x["flag_count"], reverse=True)

def build_summary(trades, flagged):
    settled = [t for t in trades if t["status"] == "SETTLED"]
    failed = [t for t in trades if t["status"] == "FAILED"]
    total_pnl = sum(t["pnl_eur"] for t in settled)
    by_desk = {}
    for t in settled:
        desk = t["desk"]
        by_desk[desk] = by_desk.get(desk, 0) + t["pnl_eur"]
    return {
        "total_trades": len(trades), "settled": len(settled), "failed": len(failed),
        "total_pnl_eur": round(total_pnl, 2), "flagged_count": len(flagged),
        "by_desk": {k: round(v, 2) for k, v in by_desk.items()},
        "generated_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

def generate_html_report(trades, flagged, summary):
    pnl_color = "#166534" if summary["total_pnl_eur"] >= 0 else "#991B1B"
    desk_rows = "".join(f'<tr><td>{desk}</td><td style="color:{"#166534" if pnl>=0 else "#991B1B"};font-weight:500">€{pnl:,.2f}</td></tr>' for desk, pnl in summary["by_desk"].items())
    flagged_rows = "".join(f'<tr style="background:{"#FEE2E2" if t["flag_count"]>=2 else "#FEF3C7"}"><td><strong>{t["id"]}</strong></td><td>{t["instrument"]}</td><td>{t["type"]}</td><td>€{t["pnl_eur"]:,.2f}</td><td><span style="background:{"#991B1B" if t["flag_count"]>=2 else "#92400E"};color:white;padding:2px 8px;border-radius:4px;font-size:11px">{"HIGH" if t["flag_count"]>=2 else "MEDIUM"}</span></td><td style="font-size:12px;color:#374151">{" · ".join(t["flags"])}</td></tr>' for t in flagged[:15])

    return f"""
    <div style="font-family: -apple-system, sans-serif; background: #F8FAFC; color: #1E293B; padding: 20px; border-radius: 8px;">
        <div style="background: #1B3A6B; color: white; padding: 24px; border-radius: 8px; display: flex; justify-content: space-between; margin-bottom: 20px;">
            <div><h1 style="margin:0;">MiddleLayer</h1><p style="margin:0; opacity:0.8;">Ops Intelligence for Fintech</p></div>
            <div style="text-align:right; font-size:12px;">Generated: {summary['generated_at']}</div>
        </div>
        <div style="display: grid; grid-template-columns: repeat(5, 1fr); gap: 16px; margin-bottom: 24px;">
            <div style="background: white; border: 1px solid #E2E8F0; padding: 16px; border-radius: 8px; text-align: center;"><div style="font-size: 24px; font-weight: bold; color:{pnl_color}">€{summary['total_pnl_eur']:,.0f}</div>Total PnL</div>
            <div style="background: white; border: 1px solid #E2E8F0; padding: 16px; border-radius: 8px; text-align: center;"><div style="font-size: 24px; font-weight: bold;">{summary['total_trades']}</div>Total Trades</div>
            <div style="background: white; border: 1px solid #E2E8F0; padding: 16px; border-radius: 8px; text-align: center;"><div style="font-size: 24px; font-weight: bold; color:#166534">{summary['settled']}</div>Settled</div>
            <div style="background: white; border: 1px solid #E2E8F0; padding: 16px; border-radius: 8px; text-align: center;"><div style="font-size: 24px; font-weight: bold; color:#991B1B">{summary['failed']}</div>Failed</div>
            <div style="background: white; border: 1px solid #E2E8F0; padding: 16px; border-radius: 8px; text-align: center;"><div style="font-size: 24px; font-weight: bold; color:#92400E">{summary['flagged_count']}</div>Flagged</div>
        </div>
        <div style="background: white; border: 1px solid #E2E8F0; padding: 20px; border-radius: 8px; margin-bottom: 20px;">
            <h2 style="margin-top:0; font-size:16px;">Flagged Trades</h2>
            <table style="width: 100%; border-collapse: collapse; text-align: left;">
                <tr style="background: #F1F5F9; font-size: 12px;"><th style="padding:10px;">ID</th><th style="padding:10px;">Instrument</th><th style="padding:10px;">Type</th><th style="padding:10px;">PnL (EUR)</th><th style="padding:10px;">Severity</th><th style="padding:10px;">Flags</th></tr>
                {flagged_rows}
            </table>
        </div>
        <div style="background: white; border: 1px solid #E2E8F0; padding: 20px; border-radius: 8px;">
            <h2 style="margin-top:0; font-size:16px;">PnL by Desk</h2>
            <table style="width: 100%; border-collapse: collapse; text-align: left;">
                <tr style="background: #F1F5F9; font-size: 12px;"><th style="padding:10px;">Desk</th><th style="padding:10px;">PnL (EUR)</th></tr>
                {desk_rows}
            </table>
        </div>
    </div>"""

# 3. EXECUTE THE PIPELINE AND DISPLAY
trades  = load_trades("trades.csv")
trades  = run_pnl_engine(trades)
flagged = detect_anomalies(trades)
summary = build_summary(trades, flagged)

# Render output right here in the notebook
display(HTML(generate_html_report(trades, flagged, summary)))
print("✅ Dashboard generated successfully!")
```
